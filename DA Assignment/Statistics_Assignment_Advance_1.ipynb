### Q1: What is a random variable in probability theory?

### Answer:
A random variable is a numerical outcome of a random phenomenon. It assigns a value (usually a number) to each possible result in a sample space, allowing mathematical treatment and analysis within probability theory.

Example: In a dice roll, the number that appears (1 to 6) is a random variable because each value has a specific probability of occurrence.

### Q2: What are the types of random variables?

### Answer:
There are two main types of random variables in statistics: discrete random variables and continuous random variables.

**Discrete Random Variable**

A discrete random variable can take only specific, countable values, such as whole numbers.

Example: The number of heads obtained when tossing a coin three times (possible values: 0, 1, 2, 3).
Example: The number of students present in a classroom (possible values: 0, 1, 2, ..., n).

**Continuous Random Variable**

A continuous random variable can take any value within a given range, and there are infinitely many possible values between any two points.

Example: The exact height of a randomly selected person (e.g., 165.3 cm, 170.75 cm, etc.).
Example: The amount of rainfall measured in a city over a year (e.g., 725.4 mm).

### Q3: Explain the difference between discrete and continuous distributions.

### Answer:
#### Discrete Distributions
- Only take on certain specific values (often countable, like integers).
- Each possible value has a specific probability, defined using a Probability Mass Function (PMF).
- The sum of the probabilities for all possible values is exactly 1.
- Examples: Number of heads in 10 coin tosses (0, 1, 2, ..., 10), rolling a die (values 1 to 6).

#### Continuous Distributions
- Can take on any value within a given range (uncountably infinite number of possibilities).
- Probabilities for exact values are 0; probabilities are assigned to intervals using a Probability Density Function (PDF).
- The total area under the curve of the PDF equals 1. Probability for an interval [a,b] is the area under the curve from a to b.
- Examples: Heights or weights of people, time taken to complete a task, rainfall amount.

### Q6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

**Answer:**
The Central Limit Theorem (CLT) states that when you take sufficiently large random samples from any population (with a finite mean and variance), the distribution of the sample means will be approximately normaleven if the original population is not normally distributed.

It is critical in statistics:
-  allows statisticians to use the normal distribution to make inferences about population parameters, even when the population itself is not normal.
- CLT is the foundation for constructing confidence intervals and hypothesis tests, which rely on the normality of sample means.
- As sample size increases (usually n  30 is considered sufficient), the sample mean's distribution becomes more closely normal, making statistical analysis more robust and reliable.

Example

Suppose you want to estimate the average height of students in a school. Instead of measuring every student, you take several random samples of 30 students each and calculate the mean height for each sample. According to CLT, the distribution of these sample means will be approximately normal, regardless of the actual distribution of heights in the school.

### Q7: What is the significance of confidence intervals in statistical analysis?

#### Answer:
Confidence intervals are crucial in statistical analysis because they provide a range of values within which the true population parameter (such as the mean or proportion) is likely to fall, given a certain level of confidence (commonly 95% or 99%).

Significance of Confidence Intervals
- Quantify Uncertainty: Confidence intervals express the uncertainty associated with a sample estimate, giving a more complete picture than a single point estimate.

- Reliability of Estimates: They indicate how reliable or precise an estimate is. A narrow interval suggests high precision, while a wide interval indicates more uncertainty.

- Statistical Significance: If a confidence interval for a difference or effect does not include zero (or another null value), it suggests the result is statistically significant at the chosen confidence level.

- Decision Making: Confidence intervals help researchers and decision-makers understand the possible range of the true effect or value, supporting better, evidence-based conclusions.


Example
If a study finds that the average test score is 80 with a 95% confidence interval of 75 to 85, it means there is a 95% chance that the true average score for the entire population lies between 75 and 85.

### Q9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print('Mean:', mean)
print('Standard Deviation:', std_dev)

# Visualization
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

**Q10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.**
Given daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

a) How would you apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval?

**Answer:**
By applying the Central Limit Theorem, we assume the sample mean of daily sales is approximately normally distributed. We calculate the mean and standard error, then use the normal distribution to estimate a 95% confidence interval for the population mean.

b) Python code to compute the mean sales and its confidence interval:

```python
import numpy as np
from scipy import stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean = np.mean(daily_sales)
std_err = stats.sem(daily_sales)  # Standard error
confidence = 0.95

margin = std_err * stats.t.ppf((1 + confidence) / 2., len(daily_sales)-1)

lower = mean - margin
upper = mean + margin

print('Mean Sales:', mean)
print('95% Confidence Interval: ({:.2f}, {:.2f})'.format(lower, upper))
```
